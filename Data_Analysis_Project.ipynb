````python
# Property name with the least number_of_reviews from the dataframe

row_least = df.loc[df['number_of_reviews'].idxmin()]
name_least = row_least['name']
print(f"Property name with least number of reviews is: {name_least}")
````

````python
# No. of properties located in hell's kitchen neighbourhood

num_hell = df[df['neighbourhood'] == "Hell's Kitchen"].shape[0]
print(f"Number of properties in Hell's Kitchen: {num_hell}")
````

````python
# Number of properties with minimum_nights more than 5

min_nights = df[df['minimum_nights'] > 5].shape[0]
print(f"Number of properties with minimum nights more than 5: {min_nights}")
````

````python
# No. of properties with reviews_per_month more than 5.0

num_rev = df[df['reviews_per_month'] > 5.0].shape[0]
print(f"Number of properties with reviews_per_month more than 5.0: {num_rev}")
````

````python
# No. of properties with last_review before 2015-12-31

df['last_review'] = pd.to_datetime(df['last_review'])
rev_2015 = df[df['last_review'] < '2015-12-31']
prop_bef_2015 = rev_2015.shape[0]

print(f"Number of properties with last_review before 2015-12-31: {prop_bef_2015}")
````

````python
# Property name with the oldest last_review with respective date

oldest_row = df.loc[df['last_review'].idxmin()]
oldest_name = oldest_row['name']
oldest_date = oldest_row['last_review']
print(f"Property name with the oldest last review : {oldest_name}")
print(f"Date of the oldest last review : {oldest_date}")
````

````python
# line graph depicting the avg price vs avg number_of_reviews

avg_price_reviews = df.groupby('number_of_reviews')['price'].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(avg_price_reviews['number_of_reviews'], avg_price_reviews['price'])
plt.title('Average Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
````
